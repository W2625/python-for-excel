```python
import pandas as pd
```

```python
pd.read_excel("xl/course_participants.xlsx")
```

```python
data=[["Mark", 55, "Italy", 4.5, "Europe"],
      ["John", 33, "USA", 6.7, "America"],
      ["Tim", 41, "USA", 3.9, "America"],
      ["Jenny", 12, "Germany", 9.0, "Europe"]
     ]
df = pd.DataFrame(data=data, columns=["name", "age", "country", "score", "continent"], 
                  index=[1001, 1000, 1002, 1003])
df
```

```python
df.info()
```

```python
df.index
```

```python
df.index.name = "user_id"
df
```

```python
# reset_index会将索引还原为普通列同时用默认索引替换当前索引
# 最终结果就和刚从Excel文件中得到的DataFrame一样
df.reset_index()
```

```python
# reset_index会将user_id还原成普通列
# set_index会将“name”列设置为索引
df.reset_index().set_index("name")
# 链式调用法，即：reset_index()会返回一个DataFrame，可以直接在这个DataFrame上调用另一个方法而无需写出中间值
```

```python
# reindex()会接管所有能够匹配新索引的行，未匹配到的索引会引入含有空值(NaN)的行
df.reindex([999, 1000, 1001, 1004])
```

```python
# 使用sort_index()按索引排序
df.sort_index()
```

```python
# 按一列或多列进行排序，使用sort_values
# 该例先按"continent"进行排序，后按"age"排序（只用一列排序即一个参数）
df.sort_values(["continent", "age"])
```

```python
df.columns
```

```python
# 列命名
df.columns.name = "properties"
df
```

```python
# 重命名rename()
df.rename(columns={"name": "First name","age": "Age"})
```

```python
# 删除某些列drop()
df.drop(columns=["name", "country"], 
        index=[1000, 1003])
```

```python
# 通过转置（transpose）DataFrame可以将行和列对调
# df.transpose()的简写
df.T
```

```python
# 更改DataFrame列的顺序（可使用前面用在索引上的reindex()方法，也可以使用loc[]直接给出所需要的列顺序通常会更直观）
df.loc[:, ["continent", "country", "name", "age", "score"]]
```

```python
# 使用标签选取数据loc[]
# 行和列都是用标量来选择，返回值也是标量
df.loc[1001, "name"]
```

```python
# 只用标量选择行或列，返回值是Series
df.loc[[1001, 1002], "age"]
```

```python
# 选取多行或多列，返回值是DataFrame
df.loc[:1002, ["name", "country"]]
```

```python
# 通过位置选取数据iloc[]
# 返回标量
df.iloc[0, 0]
```

```python
# 返回Series
df.iloc[[0, 2], 1]
```

```python
# 返回DataFrame
df.iloc[:3, [0, 2]]
```

```python
# 使用布尔索引选取数据【and —— &，or —— |，not —— ~】
# 只展示居住在美国且年龄在40岁以上的学员
tf = (df["age"] > 40) & (df["country"] == "USA")
# tf这个Series中只有True和False
tf
```

```python
df.loc[tf, :]
```

```python
# 对索引进行筛选
df.loc[df.index > 1001, :]
```

```python
# Python的列表可以使用in运算符判断是否包含某些对象，
# 若在Series中进行类似操作使用isin()
# 筛选出所有来自意大利和德国的学员
df.loc[df["country"].isin(["Italy", "Germany"])]
```

```python
# 不使用loc[]的情况下传递一整个布尔DataFrame作为参数：df[boolean_df]
# df[boolean_df]————即DataFrame只包含数字时这种语法特别有用
# 返回的DataFrame会在原DataFrame的基础上，把对应着False的地方变成NaN
# 当作以毫米为单位的年降雨量
rainfall = pd.DataFrame(data={"City1": [300.1, 100.2],
                             "City2": [400.3, 300.4],
                             "City3": [1000.5, 1100.6]})
rainfall
```

```python
rainfall < 400
```

```python
rainfall[rainfall < 400]
```

```python
# 使用MultiIndex选取数据
# 待排序的MultiIndex
df_multi = df.reset_index().set_index(["continent", "country"])
df_multi = df_multi.sort_index()
df_multi
```

```python
''' 
pandas对MultiIndex的输出进行了美化
不会为每一行数据重复输出最左端的索引级别（大洲）
只会在数据所在大洲发生改变时才进行输出
'''
df_multi.loc["Europe", :]
```

```python
# 通过多级索引选取数据需要提供一个元组作为参数
df_multi.loc[("Europe", "Italy"), :]
```

```python
# 选择性得重置一部分（即取消多索引）MultiIndex，可以为reset_index提供索引级别（从左至右，从0开始）参数
df_multi.reset_index(level=0)
```
